<h3>Statistical Inference - The Toolkit </h3>
Statistical inference is the technical process of drawing conclusions from data. 


**<span style="background-color: lightpink;"> Visualizations: Graphical Exploratory Data Analysis</span>**

- To visualizing data to uncover patterns, trends, and anomalies

**<span style="background-color: lightpink;">Quantitative Exploratory Data Analysis</span>**

**<font color='seagreen'><b>Central tendency:</b></font>**


- Mean: Average value
        
- Median: Middle value
        
- Mode: Most frequent value


**<font color='seagreen'><b>Dispersion:</b></font>**


    
- Standard deviation: Spread around the mean

           -  A high standard deviation means your data points are widely scattered; a low one means they’re tightly clustered. This helps you assess consistency.

          - Can help make decisions for standardizing and normalizing data

  
                    
- Variance: Squared spread

- Range and interquartile range (IQR): Spread between extremes or quartiles

  
        - to help detect outliers 


**<font color='seagreen'><b>Shape and Distribution:</b></font>**

- Skewness: Asymmetry of the column distributions. Can help fill in missing values.

        - Right-skewed (positive) - Tail stretches to the right; mean > median
        
        - Left-skewed (negative) - Tail stretches to the left; mean < median
        
        - Zero skewness - Symmetrical distribution
        
- Kurtosis: Peakedness or tail heaviness

- Frequency tables visualization and cross-tabulations

**<font color='seagreen'><b>Correlations:</b></font>**

**<span style="background-color: lightpink;">Hypothesis Testing------------------</span>**

- A method of making decisions or drawing conclusions about a population based on sample data. 


**<font color='seagreen'><b>1. State hypotheses:</b></font>**

Null hypothesis (H₀): Assumes no effect or no difference.

Alternative hypothesis (H₁): Suggests a real effect or difference.


**<font color='seagreen'><b>2. Choose significance level (α):</b></font>**

Common choices: 0.05 (→ 95% confidence interval), 0.01 (→ 99% CI).

This α sets the confidence level for your interval.


**<font color='seagreen'><b>3. Select test and compute statistic:</b></font>**

Based on your data type and assumptions (e.g., t-test, z-test).

At this stage, you also calculate the confidence interval for your estimate (e.g., mean difference, proportion).


**<font color='seagreen'><b>4. Calculate p-value:</b></font>**

Tells you how extreme your data is under H₀.

The CI complements this by showing the range of plausible values for your parameter.


**<font color='seagreen'><b>5. Make decision:</b></font>**

If the CI does not include the null value (e.g., 0 for mean difference), reject H₀.

If the CI does include the null value, fail to reject H₀.


**<font color='seagreen'><b>6. Interpret results:</b></font>**

The p-value tells you if the result is statistically significant.

The confidence interval tells you how big the effect might be and how precise your estimate is.
