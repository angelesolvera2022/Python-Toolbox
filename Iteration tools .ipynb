# .iterrows()

method in pandas to iterate over the rows of a DataFrame. It returns an iterator that generates a tuple for each row, where:

The first element of the tuple is the row's index.

The second element is a pandas Series object representing the row's data.

Example:

```python
for channel, followers in data[["channel_info", "followers"]].iterrows():
    print (channel, followers)
```

# iter()

- Converts an iterable (like a list,column, tuple, or range) into an iterator, which generates elements on demand.
- An iterator in Python is an object that allows you to loop through all the elements in a collection

```python
variable = iter(data["channel_info"])
```

```python
print(next(variable))
print(next(variable))
print(next(variable))
print(next(variable))
```

# enumerate()
- its also an iterable
-  function in Python, built-in tool that adds a counter (index) to any iterable and returns it as an enumerate object. It's often used in loops when you need both the index and the value of each item in the iterable

```python
data.head(2)
```

```python
for index ,channel in enumerate(data["channel_info"]):
    print (f"Index: {index}, Channel: {channel}")
    
```

```python
for index ,channel in enumerate(data["channel_info"]):
    print (index,channel)
```

```python
# can also specify where the index starts 
for index ,channel in enumerate(data["channel_info"], start = 10):
    print (f"Index: {index}, Channel: {channel}")
```

# iterate over data using chunksize = 

```python

result = []
for chunk in pd.read_csv("top_insta_influencers_data.csv", chunksize = 20):
    result.append(chunk)
print(result)
    
```

```python
result2 = []
for chunk in pd.read_csv("top_insta_influencers_data.csv", chunksize = 20):
    result2.append(sum(chunk["influence_score"]))
total = sum(result2)
print(total)
```

# map
- Applies a specified function to each item in an iterable (like a list, tuple, or set) and returns a map object, which is an iterator. It's particularly useful for transforming data in an iterable without the need for a loop.

```python
# Example: Convert a list of numbers to their squares
numbers = [1, 2, 3, 4, 5]
squares = map(lambda x: x ** 2, numbers)

# Convert the map object to a list for viewing
print(list(squares))
```

# items

- used to iterate over keys and values pairs in dictionaries. Returns a tuple of the key and value. 

```python
my_dict = {'name': 'Angeles', 'age': 25, 'city': 'San Diego'}

for key, value in my_dict.items():
    print(f"{key}: {value}")
```
